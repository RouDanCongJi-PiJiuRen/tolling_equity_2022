```python
# runs on Shirley's local machine inside a dir
# w/ all the tmcRateFileDisplayReport_<date in 2018>.xlsx files

import pandas as pd
import os

path = os.getcwd()
print(path)
```

```python
files = os.listdir('/opt/dssg-hot/data/toll_prices/pythonscripts')
files
```

```python
files_xls = [f for f in files if f[-4:] == 'xlsx']
files_xls_sorted = sorted(files_xls)
files_xls_sorted
```

```python
df = pd.DataFrame()
```

```python
for f in files_xls_sorted:
    data = pd.read_excel(f, skiprows=4)
    df = df.append(data)
```

```python
df.head()
```

```python
# want only 3 cols
dfmod = df[['Trip Fare effective Time','Trip Definition ID','Primary Rate']]
```

```python
# rename
dfmod = dfmod.rename(index=str, columns={'Trip Fare effective Time': 'Time', 'Primary Rate': 'Price'})
```

```python
# want M-F
dfmod['weekday_num'] = dfmod['Time'].apply(lambda t: t.weekday())
dfmod = dfmod[dfmod['weekday_num'].between(0, 4, inclusive=True)]
```

```python
# want 5am-7pm
dfmod['hour'] = dfmod['Time'].apply(lambda t: t.hour)
#dfmod = dfmod[dfmod['hour'].between(5, 19, inclusive=True)]
dfmod = dfmod[dfmod['hour'].between(5, 18, inclusive=True)]
```

```python
dfmod1 = dfmod.sort_values(by='Time')
```

```python
dfmod2 = dfmod1[['Time','Trip Definition ID','Price']]
```

```python
dfmod3 = dfmod2.drop_duplicates(keep=False, inplace=False)
```

```python
dfmod4 = dfmod3.pivot(index='Time', columns='Trip Definition ID', values='Price')
```

```python
dfmod4.reset_index(level=0, inplace=True)
```

```python
#dfmod.to_csv(r'concatenated_405_prices.csv')
#dfmod1.to_csv(r'concatenated_405_prices1.csv')
#dfmod2.to_csv(r'concatenated_405_prices2.csv', index = None, header = True)
#dfmod3.to_csv(r'concatenated_405_prices3.csv', index = None, header = True)
dfmod4.to_csv(r'concatenated_405_prices4.csv', index = None, header = True)
```
