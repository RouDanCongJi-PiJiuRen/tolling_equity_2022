```python
from mpl_toolkits.basemap import Basemap
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python

shape = gpd.GeoDataFrame.from_file('city10/city10.shp')
shape = shape.to_crs(epsg=4326)
print(shape.shape)
print(shape.head(10))
fig, ax = plt.subplots()
fig.set_size_inches(20, 20)
map = Basemap(projection='cyl',epsg = 4326, llcrnrlon=-123.2, llcrnrlat=46 , urcrnrlon= -120.2 , urcrnrlat=49)
map.arcgisimage(service='Canvas/World_Light_Gray_Base', verbose=True, xpixels = 1000, ypixels = 1000, dpi = 1000)#'World_Street_Map', xpixels = 1500, verbose= True)
shape.plot(column='POPHISP',ax =ax)
```
